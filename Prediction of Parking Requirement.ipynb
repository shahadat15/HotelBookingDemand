# Parking Demand Prediction
Parking demand varies over the year. Therefore if the hotel can determine the parking requirements by time of year then they can plan accordingly. They can plan for additional parking or change extra doller to park during peak seasons to limit the parking. During off season they can sub lease their parking space. 

```python
fig, ax = plt.subplots(figsize=(15,7))
d[["RequiredCarParkingSpaces","ArrivalDateMonth","ArrivalDateYear"]].groupby(["ArrivalDateMonth","ArrivalDateYear"]).count().unstack().plot(ax=ax)
ax.set_xticklabels(("aa",'January', 'March', 'May','July','September','November'))
ax.set_title('Parking Demand Trend at Differnt Time of the Month')
```

# Initial Finding
The parking demand has a definite trend over the months of a year. However the parking demand increases over the year. During August, and September 2016, the parking demand was unexpectively high. May be this variation could be predicted using other variables.
